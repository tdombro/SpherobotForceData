## Calculate the Effective Spring Constant During an Oscillation

### Modules being used

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Mon Oct  8 10:48:11 2018

@author: thomas
"""

import os, shutil
import pandas as pd
import numpy as np
import random
import matplotlib
import matplotlib.pyplot as plt
```

### Constants

```python
tstep = 5.0e-5
RADIUS = 0.15
RHO = 2.0
MASS = RHO*(4.0/3.0)*np.pi*RADIUS**3
print('mass = ',MASS)
St = 0.78
FREQ = 10.0
OMEGA = 2.0*np.pi*FREQ
AMP = RADIUS/St
FIGNUM = 0
```

### Data Table Names

```python
ForceString = 'fdRe30.0St0.78.txt'
PositionString = 'pdRe30.0St0.78.txt'
```

```python
def main(ForceString, PositionString):
    global tstep, MASS, RADIUS, RHO, AMP, OMEGA
    #Create 'data' dataframe. Must include input values
    '''Read and import Force data file'''
    #Columns used by data.'NameofColumnHere'
    fd = pd.read_csv(ForceString,delimiter = ' ',names=['Fx0','Fy0','Fx1','Fy1','time'])
    fd = fd.dropna()
    fd = fd.sort_values(by=['time'])
    '''Read and import Position data file'''
    #Columns used by data.'NameofColumnHere'
    posd = pd.read_csv(PositionString,delimiter = ' ',names=['x0','y0','x1','y1','curr','des','time'])
    posd = posd.dropna()
    posd = posd.sort_values(by=['time'])
    posd.y1 += 0.75
    
    #Add new Columns to force database
    fd['NetFcorr1'], fd['NetFcorr0'] = 0.0,0.0
    fd['Keff1'], fd['Keff0'] = 0.0,0.0
    #display(fd)
    #Input Initial correction forces
    fd.at[0,'NetFcorr1'] = fd.at[0,'Fy1']
    fd.at[0,'NetFcorr0'] = fd.at[0,'Fy0']
    #Input Initial Keff values
    fd.at[0,'Keff1'] = abs(fd.at[0,'Fy1']/posd.at[0,'y1'])
    fd.at[0,'Keff0'] = abs(fd.at[0,'Fy0']/posd.at[0,'y0'])
    
    SSAnalysis(posd,fd)
    
    LSAnalysis(posd,fd)
    
    return

#------------------------__END MAIN__----------------------------
```

### Analysis Script for Small Sphere

```python
def SSAnalysis(posd,fd):
    global tstep, MASS, RADIUS, RHO, AMP, OMEGA, FIGNUM
    #This function will be used to analyze the position data as well as the force data
    
    
    #Calculate the effective spring constant for quarter oscillations
    tosc = 0.025
    NoscMax = int(3.0/tosc)
    #Create Numpy array of net forces per quarter oscillation
    NetForce1 = np.zeros((2,NoscMax))
    oscNum = np.arange(1,NoscMax+1,1)
    Ypos1 = np.zeros(NoscMax)
    for idx in range(NoscMax):
        maxTime = (idx+1)*tosc
        #print('maxTime = ',maxTime)
        NetForce1[0,idx] = np.sum(fd.Fx1[(fd.time <= maxTime) & (fd.time > maxTime - tosc)])
        NetForce1[1,idx] = np.sum(fd.Fy1[(fd.time <= maxTime) & (fd.time > maxTime - tosc)])
        Ypos1[idx] = posd.at[int(maxTime/tstep)-1,'y1']
        #print('NetForce[%i]'%idx)
        #print(NetForce1[:,idx])
    
    #Plot Net Force per quarter oscillation for whole sim
    #Alongside, plot position of sphere at each quarter oscillation
    fig2 = plt.figure(num=FIGNUM, figsize=(8,4),dpi=120)
    ax = fig2.add_subplot(121)
    ax.set_xlabel('Quart Osc Num',fontsize=12)
    ax.set_ylabel('Keff1 (N)',fontsize=12)
    ax.plot(oscNum*tosc,NetForce1[1,:]/AMP)
    ax.scatter(oscNum*tosc,NetForce1[1,:]/AMP,s=9)
    ax.axis([tstart,tend,np.amin(NetForce1)/AMP,np.amax(NetForce1)/AMP])
    ax2 = fig2.add_subplot(122)
    ax2.set_xlabel('Quart Osc Num',fontsize=12)
    ax2.set_ylabel('Ypos1 (m)',fontsize=12)
    ax2.plot(oscNum*tosc,Ypos1)
    ax2.scatter(oscNum*tosc,Ypos1,s=9)
    #ax2.axis([1.0,1.1,-0.15,0.15])
    fig2.tight_layout()
    plt.show()
    fig2.clf()
    FIGNUM += 1
        
    return
```

### Analysis Script for Large Sphere

```python
def LSAnalysis(posd,fd):
    global tstep, MASS, RADIUS, RHO, AMP, OMEGA, FIGNUM
    #This function will be used to analyze the position data as well as the force data
    #Calculate the effective spring constant as a function of time
    for idx in range (1,len(posd.time)):
        #Calculate the Net Correction Force
        fd.at[idx,'NetFcorr0'] = fd.at[idx,'Fy0'] + fd.at[idx-1,'NetFcorr0']
        #Calculate Keff
        fd.at[idx,'Keff0'] = abs(fd.at[idx,'Fy0']/posd.at[idx,'y0'])
    NetForce = np.sum(fd.Fy0)
    print('Net Force = ',NetForce)
    print('Dataset Net Force = ',fd.at[len(fd.time) - 1,'NetFcorr0'])
    
    tstart, tend = 2.0, 2.1
    #Plot Keff vs time
    fig2 = plt.figure(num=FIGNUM, figsize=(12,4),dpi=120)
    ax = fig2.add_subplot(131)
    ax.set_xlabel('time (s)')
    ax.set_ylabel('Keff0 (N/m)')
    ax.plot(posd.time,fd.Keff0)
    ax.axis([tstart,tend,0,1.0e4])
    ax2 = fig2.add_subplot(132)
    ax2.set_xlabel('time (s)')
    ax2.set_ylabel('NetFcorr0 (N)')
    ax2.plot(fd.time,fd.NetFcorr0)
    #ax2.axis([tstart,tend,np.amin(fd.NetFcorr0[(fd.time <= tend) & (fd.time >= tstart)]),np.amax(fd.NetFcorr0[(fd.time <= tend) & (fd.time >= tstart)])])
    ax3 = fig2.add_subplot(133)
    ax3.set_xlabel('time (s)')
    ax3.set_ylabel('Fcorr0 (N)')
    ax3.plot(fd.time,fd.Fy0)
    ax3.axis([tstart,tend,np.amin(fd.Fy0[(fd.time <= tend) & (fd.time >= tstart)]),np.amax(fd.Fy0[(fd.time <= tend) & (fd.time >= tstart)])])
    fig2.tight_layout()
    plt.show()
    fig2.clf()
    FIGNUM += 1
        
    return
```
